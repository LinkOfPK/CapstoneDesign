```python
import requests
import json

# OpenWeatherMap API í‚¤
API_KEY = "13a24ecc3a2785b9652b8816b02a8e55"

def get_lat_lon_from_ip():
    """IP ê¸°ë°˜ìœ¼ë¡œ ìœ„ë„ì™€ ê²½ë„, ë„ì‹œëª… ë°˜í™˜"""
    try:
        response = requests.get("https://ipinfo.io/json", timeout=5)
        data = response.json()
        lat, lon = map(float, data["loc"].split(","))
        city = data.get("city", "ì•Œ ìˆ˜ ì—†ìŒ")
        print(f"ğŸ“ ìë™ ìœ„ì¹˜ ê°ì§€: {city} (ìœ„ë„ {lat}, ê²½ë„ {lon})")
        return lat, lon, city
    except Exception as e:
        print("âŒ IP ê¸°ë°˜ ìœ„ì¹˜ ì¡°íšŒ ì‹¤íŒ¨:", e)
        return None, None, None

def get_lat_lon(city_name):
    """ì‚¬ìš©ì ì…ë ¥ ë„ì‹œëª…ì„ ìœ„ë„/ê²½ë„ë¡œ ë³€í™˜"""
    geo_url = f"https://api.openweathermap.org/geo/1.0/direct?q={city_name}&limit=1&appid={API_KEY}"
    response = requests.get(geo_url)
    
    if response.status_code == 200 and response.json():
        data = response.json()[0]
        return data["lat"], data["lon"]
    else:
        print("âŒ ì…ë ¥í•œ ë„ì‹œì˜ ìœ„ì¹˜ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return None, None

def get_weather(lat, lon):
    """ìœ„ë„ì™€ ê²½ë„ë¡œ ë‚ ì”¨ ë°ì´í„° ê°€ì ¸ì˜¤ê¸°"""
    weather_url = f"https://api.openweathermap.org/data/2.5/weather?lat={lat}&lon={lon}&appid={API_KEY}&units=metric"
    response = requests.get(weather_url)

    if response.status_code == 200:
        return response.json()
    else:
        print("âŒ ë‚ ì”¨ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ëŠ” ë° ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤.")
        return None

# âœ… ìœ„ì¹˜ ì¡°íšŒ ì‹œë„: ìë™ â†’ ì‹¤íŒ¨ ì‹œ ìˆ˜ë™
lat, lon, city = get_lat_lon_from_ip()

# ìë™ ì¡°íšŒ ì‹¤íŒ¨ ì‹œ ì‚¬ìš©ìì—ê²Œ ë„ì‹œëª…ì„ ì…ë ¥ë°›ê¸°
if lat is None or lon is None:
    city = input("ğŸŒ ìë™ ìœ„ì¹˜ ì¡°íšŒ ì‹¤íŒ¨! ì§ì ‘ ë„ì‹œëª…ì„ ì…ë ¥í•˜ì„¸ìš”: ")
    lat, lon = get_lat_lon(city)

# ë‚ ì”¨ ë°ì´í„° ì¡°íšŒ
if lat and lon:
    weather_data = get_weather(lat, lon)

    if weather_data:
        # ğŸŒ¤ï¸ í…ìŠ¤íŠ¸ í‘œí˜„ (UIìš©)
        weather = weather_data["weather"][0]["main"]
        description = weather_data["weather"][0]["description"]

        # ğŸŒ¡ï¸ ê°ì„± ë²¡í„°ì— í•„ìš”í•œ ì •ëŸ‰ ë°ì´í„° ì¶”ì¶œ
        temperature = weather_data["main"]["temp"]                      # í˜„ì¬ ê¸°ì˜¨
        feels_like = weather_data["main"]["feels_like"]                # ì²´ê° ê¸°ì˜¨
        humidity = weather_data["main"]["humidity"]                    # ìŠµë„ (%)
        cloudiness = weather_data["clouds"]["all"]                     # êµ¬ë¦„ ì–‘ (%)
        wind_speed = weather_data["wind"]["speed"]                     # í’ì† (m/s)
        visibility = weather_data.get("visibility", 10000)             # ê°€ì‹œê±°ë¦¬ (m)

        # ğŸ•’ ì‹œê°„ ê¸°ë°˜ ë°ì´í„°
        sunrise = weather_data["sys"]["sunrise"]                       # ì¼ì¶œ ì‹œê°„ (UNIX timestamp)
        sunset = weather_data["sys"]["sunset"]                         # ì¼ëª° ì‹œê°„ (UNIX timestamp)
        current_time = weather_data["dt"]                              # í˜„ì¬ ì‹œê° (UNIX timestamp)

        # ğŸ–¨ï¸ ì¶œë ¥ í™•ì¸
        print(f"\nğŸ“ [ {city} ] ì˜ ë‚ ì”¨ ì •ë³´")
        print(f"ğŸŒ¤ï¸ ìƒíƒœ: {weather} ({description})")
        print(f"ğŸŒ¡ï¸ ê¸°ì˜¨: {temperature}Â°C")
        print(f"ğŸ¥¶ ì²´ê° ê¸°ì˜¨: {feels_like}Â°C")
        print(f"ğŸ’§ ìŠµë„: {humidity}%")
        print(f"â˜ï¸ êµ¬ë¦„ ì–‘: {cloudiness}%")
        print(f"ğŸŒ¬ï¸ í’ì†: {wind_speed} m/s")
        print(f"ğŸ‘ï¸ ê°€ì‹œê±°ë¦¬: {visibility} m")
        print(f"ğŸŒ… ì¼ì¶œ: {sunrise} | ğŸŒ‡ ì¼ëª°: {sunset} | ğŸ•’ í˜„ì¬ ì‹œê°„: {current_time}")
    else:
        print("âŒ ë‚ ì”¨ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ëŠ” ë° ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤.")
else:
    print("âŒ ìœ„ì¹˜ ì •ë³´ë¥¼ í™•ì¸í•  ìˆ˜ ì—†ì–´ ë‚ ì”¨ ì¡°íšŒë¥¼ ì§„í–‰í•  ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
```

```python
import json
import os
# íŒŒì¼ë¡œ ì €ì¥ (JSON)


# íŒŒì¼ ê²½ë¡œ
file_path = "weather_data.json"

# JSON íŒŒì¼ë¡œ ì €ì¥
try:
    with open(file_path, "w") as f:
        json.dump(weather_data, f, indent=4)
    
    # ì €ì¥ ì„±ê³µ ì—¬ë¶€ í™•ì¸
    if os.path.exists(file_path):
        print(f"âœ… íŒŒì¼ì´ ì„±ê³µì ìœ¼ë¡œ ì €ì¥ë˜ì—ˆìŠµë‹ˆë‹¤: {file_path}")
    else:
        print("âš ï¸ ì €ì¥ì— ì‹¤íŒ¨í–ˆìŠµë‹ˆë‹¤. íŒŒì¼ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")

except Exception as e:
    print(f"âŒ ì˜¤ë¥˜ ë°œìƒ: {e}")
```
